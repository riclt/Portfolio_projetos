# Descobrindo o pandas
Com o intuito de facilitar, otimizar e automatizar nossas tarefas com análise de dados, onde normalmente são feitas através do Excel, conheci e apresento a vocês essa ferramenta incrível chamada: **Pandas**.

É praticamente impossível não associar esse nome ao urso. Mas a verdade é que o Pandas é uma biblioteca poderosa para manipulação e análise de dados através da linguagem Python.

O Pandas é uma biblioteca *open source* de Python e uma das suas grandes facilidades é a apresentação tabular dos dados, o que tem muita semelhança com a ferramenta Excel que já conhecemos e utilizamos.

Vamos conhecer um pouco das muitas utilizações possíveis com a biblioteca Pandas e como pude aplicá-la de maneira prática e simples.







# Importando o Pandas

Através do nosso ambiente de programação (colab), podemos importar diretamente a biblioteca Pandas através do comando ***import***.

Para otimizar um pouco mais a nossa vida, podemos colocar um apelido na biblioteca Pandas, denominado "pd", e assim facilitar nosso fluxo de trabalho no código.

```python
# import pandas
import pandas as pd
```

# Planilhas em Excel para o Pandas

Colocando em prática a utilização da biblioteca Pandas como ferramenta de trabalho, vamos utilizar como exemplo uma planilha simples do Excel com dados reais e pessoais de vendas anuais.

Considerando que a planilha foi carregada ao seu ambiente de programação, através do *Google Drive* ou dos *uploads* dos seus arquivos, podemos fazer a leitura dos dados.

Para a leitura dos dados do arquivo em Excel utilizamos a função: ***pd.read_excel***. Além disso, podemos associar esse arquivo a uma variável.

```python
# importar planilha do excel
planilha = pd.read_excel("/content/anual_vendas_python.xlsx")
```

Após a leitura do arquivo, vamos executar e visualizar o carregamento da nossa planilha.

```python
# executar planilha
planilha
```

# Acessar uma aba específica da planilha
Vamos considerar a necessidade ao acesso somente a aba **Total de Vendas** da planilha.

Basta colocar o comando ***planilha[" Total de vendas "]***.

Lembrando que o **Python** é sensível a variações de caracteres, como por exemplo, *"Total de Vendas"* é diferente de *"total de vendas"*.

```python
# visualizando a aba Total de Vendas
planilha[" Total de vendas "]
```

Podemos visualizar todas as abas existentes na planilha utilizando o comando ***planilha.keys()***.

Este comando é bastante útil para conhecer todas abas da planilha e carregar a aba necessária no momento da manipulação dos dados.

```python
# verificar abas da planilha
planilha.keys()
```

# Manipulando valores da planilha

Posso, através do ***pandas***, trabalhar com inúmeras operações para manipulação dos dados, como: soma, média, desvio padrão, entre muitas outras.

Para somar os valores do total de vendas, por exemplo, basta utilizar o comando ***sum***.

```python
# Somar o total de vendas
planilha[" Total de vendas "].sum()
```

Posso descobrir minha média anual de vendas através do comando ***mean***.

```python
# média anual das vendas
planilha[" Total de vendas "].mean()
```

E ainda assim, posso utilizar um dos grandes recursos da estatística na distribuição de dados que estão em curso, que é o desvio padrão.

Basta utilizar o comando ***std*** *(Standard Deviation)*.

```python
# Calcular o desvio padrão do total de vendas
planilha[" Total de vendas "].std()
```

# Gerar gráficos com os dados da planilha

Outro aspecto valiosíssimo descoberto com o pandas, é sua a capacidade de plotar gráficos de maneira simples, prática, fácil e eficaz.

Como forma de exemplo e prática, vamos plotar um gráfico de linhas da relação: Ano / total de vendas.

```python
# plotar gráfico de linhas
planilha.plot('Ano', ' Total de vendas ')
```
